```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
categories=np.array(["A","B","C","D","E"])
values=np.array([23,56,41,62,19])
plt.subplot(2,2,1)
plt.bar(categories,values)
plt.title("Bar Graph")
x=np.array([12,17,21,18,14,13,16,9,12,15,19,11,14,16,20,18,15,13,16,11,10])
plt.subplot(2,2,2)
plt.hist(x)
plt.title("Histogram")
y=np.array([30,25,15,30])
lbl=["Apple","Banana","Orange","Mango"]
plt.subplot(2,2,3)
plt.pie(y,labels=lbl)
plt.title("Pie Plot")
x1=np.array([1,2,3,4,5])
y1=np.array([2,5,3,6,4])
plt.subplot(2,2,4)
plt.scatter(x1,y1)
plt.title("Scatter plot")
plt.subplots_adjust(top=0.92, bottom=0, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.50)
plt.show()
```
